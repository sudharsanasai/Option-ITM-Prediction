```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

badf = pd.read_csv('data/badf.csv')

badf['date'] = pd.to_datetime(badf['date'])

badf = badf[badf['date']<'2020-01-01']

pd.set_option('display.max_columns', None)
badf.describe()


# for threshold in [0, 0.0005, 0.001, 0.002, 0.004, 0.007, 0.01]:
#     print(f'Threshold= {threshold}')
#     for col in badf.columns:
#         if 'target_' in col:
#             label = col.replace('target_', 'label_')
#             badf[label] = (badf[col]>threshold).astype(int)
#             print(badf[label].value_counts())

threshold = 0.001
badf['label'] = (badf['target_5d']>threshold).astype(int)
            
badf.drop(columns=['open', 'high', 'low', 'close'], inplace=True)

badf.set_index('date',inplace=True)
# badf['label'] = badf['label_5d']
# badf = badf.drop(columns = ['target_1d','target_5d','target_10d','target_21d','label_1d','label_10d','label_21d','label_5d'])
badf = badf.drop(columns = ['target_1d','target_5d','target_10d','target_21d'])
badf = badf.drop(columns = [column for column in badf.columns if 'fb' in column])
badf = badf.drop(columns = [column for column in badf.columns if 'tsla' in column])

for column in badf.columns:
    badf[column] = badf[column].fillna(badf[column].median())
```

## Decision Tree 

```python
train_size = 4586
val_size = 100

train_df = badf[4100:train_size]
test_df = badf[train_size:train_size+val_size]

train_df_X = train_df.drop('label',axis=1)
train_df_Y = train_df['label']
test_df_X = test_df.drop('label',axis=1)
test_df_Y = test_df['label']

from sklearn.tree import DecisionTreeClassifier
import sklearn.metrics
```

```python
model = DecisionTreeClassifier(criterion='entropy').fit(train_df_X,y = train_df_Y)
model.predict(test_df_X)
```

```python
feature_importance = pd.Series(dict(zip(train_df_X,model.feature_importances_))).sort_values(ascending=False)
feature_importance.head(50)
```

# Baseline Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import confusion_matrix
```

```python
train_size = 4586
val_size = 100


train_df = badf[:-300]
train_df = train_df
test_df = badf[-300:]
test_df = test_df

train_df_X = train_df.drop('label',axis=1)
train_df_Y = train_df['label']
test_df_X = test_df.drop('label',axis=1)
test_df_Y = test_df['label']
```

```python
model = DecisionTreeClassifier(random_state=42)
model.fit(train_df_X,train_df_Y)
y_pred_prob = model.predict_proba(test_df_X)
y_pred = model.predict(test_df_X)
roc_auc_score(test_df_Y,y_pred_prob[:,1])
```

# Random Forest Feature Importance
